# Defining Template 1 : **Short & Supportive Response**

We designed it to provide a short response to the user’s emotional input. It predicts the user’s primary emotion from their text and delivers concise advice in a few sentences. The focus is on immediate reassurance while using a calm and comforting tone to make the user feel understood and supported. This template is ideal for situations where the user wants quick guidance without deep analysis.

## **Advantages**
- **Quick**  
- **Easy to read**  
- **Calming tone**  
- **Ideal for users seeking immediate reassurance**

## **Disadvantages**
- **Lacks detailed emotional reasoning**  
- **Provides minimal actionable steps**  
- **Less educational**  
- **Does not explore underlying causes of the emotion**


```python
def generate_response_template1(text):
    prompt = f"""
    You are AURA, an emotional health assistant.

    The user wrote:
    "{text}"

    Identify the user's emotion. Only choose from these five: sadness, anger, love, fear, joy.
    Then give SHORT advice in 2–3 sentences.
    Make the tone calm, supportive, and simple. 

    IMPORTANT: Do not ask the user any follow-up questions. Give a single response only.
    """

    response = client.chat.completions.create(
        model="llama-3.1-8b-instant",
        messages=[{"role": "user", "content": prompt}]
    )

    return response.choices[0].message.content
```

# Defining Template 2 : **Detailed & Comprehensive Response**

Designed to offer a detailed well-structured response that goes beyond simply identifying the user’s emotion. It predicts the emotion and explains why the user might be feeling that way then provides comprehensive advice to help them manage their feelings. The guidance includes a grounding technique, behavioral suggestions, and a positive coping perspective that is delivered in a warm and emotionally reassuring tone. This template is intended for users who want deeper insight into their emotional state and practical strategies to improve their well-being.

## **Advantages**
- **Highly detailed**
- **Actionable advice**
- **Educational and reflective**
- **Supports deeper emotional understanding**

## **Disadvantages**
- **Longer output may overwhelm some users**
- **Takes more time to read**

```python
def generate_response_template2(text):
    prompt = f"""
    You are AURA, an advanced emotional well-being guide.

    The user wrote:
    "{text}"

    Your task:
    1. Detect the user's emotion. Only choose from these five: sadness, anger, love, fear, joy.
    2. Explain WHY they might be feeling this way.
    3. Give detailed advice including:
       - a grounding technique
       - a behavioral suggestion
       - a positive coping perspective
    4. Keep the tone warm, human, and emotionally safe.

    IMPORTANT: Do not ask the user for further input. Provide the complete detailed advice in this single response.
    """

    response = client.chat.completions.create(
        model="llama-3.1-8b-instant",
        messages=[{"role": "user", "content": prompt}]
    )

    return response.choices[0].message.content
```

Prompting user to enter a sentence on how they are feeling to generate responses

```python
console = Console()
user_input = input("Enter how you're feeling: ")


template1_response = generate_response_template1(user_input)
template2_response = generate_response_template2(user_input)

# Template 1
console.print("\nTEMPLATE 1: SHORT SUPPORTIVE RESPONSE", style="bold green")
console.print(Panel.fit(f"[bold yellow]User Input:[/bold yellow] {user_input}\n\n{template1_response}", border_style="green"))

# Template 2
console.print("\nTEMPLATE 2: DETAILED COMPREHENSIVE RESPONSE", style="bold blue")
console.print(Panel.fit(f"[bold yellow]User Input:[/bold yellow] {user_input}\n\n{template2_response}", border_style="blue"))
```

After comparing both response templates, we concluded that **template 2** provides the most value to the user in terms of emotional clarity, depth, and support: 

## Template Comparison


| Aspect           | Template 1                     | Template 2                          |
|-----------------|---------------------------------|-------------------------------------|
| **Purpose**      | Quick reassurance               | In-depth guidance                   |
| **Emotion**      | Detects primary emotion         | Detects emotion + explains why      |
| **Advice**       | Brief 2–3 sentences             | Detailed with techniques            |
| **Time**         | Short                           | Longer                              |



### **Why Template 2 Was Selected**

Template 2 is chosen for our case because it provides a more comprehensive and meaningful experience for the user. While Template 1 offers quick reassurance and short advice, Template 2 goes further by not only identifying the user’s emotion but also explaining potential reasons behind it and offering actionable strategies for emotional regulation.

This deeper level of guidance ensures that users gain both emotional insight and practical tools to manage their feelings. As a result, Template 2 becomes the most valuable option for users seeking thoughtful, supportive, and development-oriented responses rather than brief comfort.
